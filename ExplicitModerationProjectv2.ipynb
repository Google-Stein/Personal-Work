```python
def contains_banned_term(prompt, wordlist_path="banned_terms.txt"):
    with open(wordlist_path, "r", encoding="utf-8") as f:
        banned_terms = [line.strip().lower().strip(",\"'") for line in f if line.strip()]

    prompt = prompt.lower()
    matches = []

    for term in banned_terms:
        # Match full word only using word boundaries
        pattern = r"\b" + re.escape(term) + r"\b"
        if re.search(pattern, prompt):
            matches.append(term)

    return matches


def interactive_prompt_check(prompt, harm_map, inno_map, adj_map, wordlist_path="banned_termsv3.txt"):
    harm_score = getPromptHarmScore(prompt, harm_map, inno_map, adj_map)
    banned_matches = contains_banned_term(prompt, wordlist_path)

    is_flagged = harm_score[0] >= 5 or bool(banned_matches)

    print("\n--- Moderation Check Result ---")
    print(f"Prompt: {prompt}")
    print(f"Harm Score: {harm_score}")
    print(f"Harm Label: {'Harmful' if is_flagged else 'Harmless'}")
    if banned_matches:
        print(f"Banned Terms Matched: {banned_matches}")
    else:
        print("No banned terms matched.")
    print("-------------------------------\n")


user_input = input("Enter a prompt to evaluate: ")
interactive_prompt_check(user_input, harmDict, innoTerms, adjTerms)
```
